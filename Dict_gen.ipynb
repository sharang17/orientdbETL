```python
import pandas as pd 
import numpy as np
from itertools import chain
from collections import defaultdict  
import pickle
```

```python
paysim=pd.read_csv('sample_paysim.csv')
```

```python
paysim.head()
```

```python
index=paysim['Unnamed: 0']
paysim=paysim.drop('Unnamed: 0',axis=1)
paysim['Trans_ID']=index
paysim=paysim.set_index('Trans_ID')
```

```python
paysim.head()
```

```python
paysim.columns
```

```python
orig_df=paysim[['nameOrig','type']].reset_index().set_index('nameOrig')
dest_df=paysim[['nameDest','type']].reset_index().set_index('nameDest')
trans_df=paysim[['nameOrig','nameDest','newbalanceDest','newbalanceOrig','oldbalanceDest','oldbalanceOrg','type']]
```

```python
trans_dict=trans_df.to_dict('index')
orig_dict=orig_df.to_dict('index')
dest_dict=orig_df.to_dict('index')
```

```python
with open('trans_dict.pickle', 'wb') as handle:
    pickle.dump(trans_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('orig_dict.pickle','wb') as handle:
    pickle.dump(orig_dict,handle,protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('dest_dict.pickle','wb') as handle:
    pickle.dump(orig_dict,handle,protocol=pickle.HIGHEST_PROTOCOL)
```
